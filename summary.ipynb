```python
from transformers import T5Tokenizer,T5ForConditionalGeneration
model_name="t5-small"
tokenizer=T5Tokenizer.from_pretrained(model_name)
model=T5ForConditionalGeneration.from_pretrained(model_name)
```

```python

input_text="""In a small village surrounded by hills and rivers, there lived a boy named Aarav. He wasn’t like the other children who spent their evenings playing by the riverbank or chasing butterflies in the fields. Aarav loved to look at the night sky. Every night, after finishing his chores, he would lie on the cold roof of his house, watching the stars glitter like diamonds scattered across black velvet.
His mother used to tell him, “Aarav, dreams are beautiful, but don’t get lost in them.” But Aarav would smile and reply softly, “Amma, I don’t want to get lost — I just want to find where they lead.”
Aarav’s fascination grew as he got older. He began to read old books from the school library about astronomy, space, and the universe. He made his own telescope using discarded lenses and pipes. Most people in the village thought he was wasting time. They would laugh and say, “Stars are too far, boy. You should look at the fields, not the sky.”
But Aarav didn’t stop.
One evening, a meteor shower was predicted. Aarav stayed awake all night, eyes wide open, waiting for the first streak of light. When he finally saw it — a blazing trail cutting through the darkness — he made a wish:
“One day, I’ll go beyond those stars.”
Years passed. Aarav studied hard, earning a scholarship to a city college. Life wasn’t easy. He worked part-time, studied late nights, and sometimes went hungry just to save money for books. But every time he felt like giving up, he would look up at the faint stars above the polluted city sky and whisper, “Not yet.”
After years of struggle, Aarav became an aerospace engineer. When the national space agency announced a mission to design a satellite that could study deep space, Aarav poured his heart into it. His designs were innovative — born not from textbooks but from years of stargazing and dreaming.
The day the satellite was launched, Aarav stood among scientists, watching the rocket rise through the clouds. As it vanished into the endless blue, tears filled his eyes. For a moment, he wasn’t the engineer everyone admired — he was still that small boy on the rooftop, reaching out to the stars.
Months later, the satellite sent its first images of distant galaxies — swirling lights billions of years old. Aarav stared at them on his computer screen and smiled. Somewhere deep inside, he heard his mother’s voice:
“Dreams are beautiful, but don’t get lost in them.”
And he whispered back,
“Amma, I didn’t get lost. I just followed the light.”
"""
input="summarize:"+input_text

tokens=tokenizer.encode(input,return_tensors="pt",max_length=512,truncation=True)
summary_ids=model.generate(tokens,max_length=80,min_length=20,num_beams=4,length_penalty=2.0,early_stopping=True)
summary=tokenizer.decode(summary_ids[0],skip_special_tokens=True)
print(summary)

```
